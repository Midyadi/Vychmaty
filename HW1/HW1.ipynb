```python
import seaborn as sns
```

Задание 1

```python
matrix = np.random.normal(size = (1000, 1000))
matrix_symm = matrix + matrix.T
first_eigenvalues = np.linalg.eigvalsh(matrix_symm)
sns.histplot(first_eigenvalues, bins=np.linspace(-110,110,31), color = 'magenta')

plt.title('Eigenvalues histogram')
plt.xlabel('value')
plt.ylabel('count')
plt.show()
```

```python
bins=np.linspace(-110, 110, 31)

bin_values = np.zeros((50, 30))

for i in range(50):
    matrix = np.random.normal(size = (1000,1000))
    matrix_symm = matrix + matrix.T 
    eigenvalues= np.linalg.eigvalsh(matrix_symm)

    bin_vals, edges = np.histogram(eigenvalues, bins)
    bin_values[i] = bin_vals


eig_mean = np.mean(bin_values, axis=0)
eig_std = np.std(bin_values, axis=0)

plt.bar((bins+3.5)[:-1], eig_mean, width=6, color='r', yerr=eig_std)
plt.title('Eigenvalues histogram with errorbars')
plt.xlabel('value')
plt.ylabel('count')
plt.show()
```

Задание 2

```python
dsts = np.zeros((500,999))
for i in range(500):
    matrix = np.random.normal(size = (1000, 1000))
    matrix_symm = matrix + matrix.T
    eigenvalues = np.sort(np.linalg.eigvalsh(matrix_symm))
    dst = np.diff(eigenvalues)
    dsts[i] = dst

sns.histplot(dsts.flat, bins=np.linspace(0,3,51), color='cyan', stat="density")
plt.title("Distance between eigenvalues (N = 1000, m = 500)")
plt.xlabel('distance $\\rho$')
plt.ylabel('density $f(\\rho)$')
plt.show()
```

Задание 3

По гистограммам видно, как плотность вероятности зависит от расстояния между собственными числами:

Чем больше $N$, тем выше, уже и ближе к нулю пик, и наоборот.

Лучше всего к данным графикам подходит распределение Рэлея: 

$f(\rho; \sigma) = \dfrac{\rho}{\sigma^2}\exp(-\dfrac{\rho^2}{2\sigma^2})$, где $\rho$ - расстояние между собственными числами, а $\sigma$ - параметр распределения, который в свою очередь будет зависить от $N$ (причём чем больше $N$, тем меньше $\sigma$).

# Задание 1

```python
import numpy as np

table = np.array([np.linspace(-1.3,1.4,10), [45.84,24.70,13.69,8.02,5.37,4.99,6.81,10.62, 15.13,17.15]])
```

Поиск коэффициентов интерполяционного многочлена по некоторому количеству заданных точек $x_i$ и значений функции в них $y_i$  можно свести к решению матричного уравнения следующего вида:

$$
        \begin{bmatrix}
            1 & x_0 & \dots & x_0^n\\
            1 & x_1 & \dots & x_1^n\\
            \vdots & \ddots & \ddots & \vdots\\
            1 & x_n & \dots & x_n^n
        \end{bmatrix}
        \begin{bmatrix}
            a_0\\
            a_1\\
            \vdots\\
            a_n
        \end{bmatrix}
        = 
        \begin{bmatrix}
            y_0\\
            y_1\\
            \vdots\\
            y_n
        \end{bmatrix}
$$

В кратком виде запишем $Xa = y$

По заданию будем пытаться решить его матодом обратной матрицы. Суть метода - в непоредственном нахождении обратной матрицы $X^{-1}$ и домножении уравнения на неё, тогда можем найти столбец коэффициентов:

$$a = X^{-1}y$$

**Данный метод применим только в том случае, если определитель матрицы $X$ не равен нулю**

В нашем случае это выполняется, так как наша матрица - **матрица Вандермонда**.

```python
# Находим X^{-1}
X = np.vander(table[0], increasing=True)
X_rev = np.linalg.inv(X)

# Находим столбец коэффициентов
a_coef = X_rev @ table[1]

# Проверяем что многочлен с найденными коэффициентами проходит через все зааднные точки
np.allclose(X@a_coef, table[1])
```

```python
import matplotlib.pyplot as plt

xs = np.linspace(-1.3,1.4, 1000)
X_plt = np.vander(xs, 10, increasing=True)

ys = X_plt @ a_coef

plt.plot(xs, ys, color='navy', label="interpolation")
plt.scatter(table[0], table[1], color='crimson', label='data')
plt.legend()
plt.show()
```

# Задание 2

Аналитическое интегрирование многочлена на отрезке тривиально, итоговый результат: $$I = a_0(b-a)+\dfrac{a_1}{2}(b^2-a^2)+\cdots+\dfrac{a_n}{n+1}(b^{n+1}-a^{n+1})$$

В нашем случае $b=1.3, a=-1.4$, коэффициенты $a_i$ мы нашли в прошлом задании

```python
b=1.3
a=-1.4
n=9
der = np.arange(1,n+2)
b_s = np.array([b**i for i in range(1,n+2)])
a_s = np.array([a**i for i in range(1,n+2)])
I_real = np.sum(a_coef/der*(b_s-a_s))

I_real
```

Метод правых прямоугольников - метод интегрирования, при котором в формуле метода прямоугольников значение функции берётся с правого края прямоугольника: $$\int_a^bf(x)dx \approx f(b)(b-a)$$

```python
h = 0.1 # Данный по условию шаг интегрирования

xs = np.linspace(a,b, int((b-a)/h)+1)

ys = np.vander(xs, N=10, increasing=True) @ a_coef

I_chisl = np.sum(ys[1:])*h

print(f'Точное значение: I = {I_real}, численно посчитанное: I = {I_chisl}')
```

Численно посчитанное значение меньше аналитического. Это вполне ожидаемо, так как использовался метод правых прямоугольников, а интегрируемая функция "в основном" убывает.

Это значит, что большинство прямоугольников расположены целиком под функцией, и остаются непосчитанные "пустоты".

Оценку погрешности проведём по формуле $\Delta I = \dfrac{(b-a)h}{2}M_1$, где $M_1=\max\limits_{a \leq x \leq b}|f'(x)|$

В нашем случае видно по графику, построенному ранее, что $M_1$ достигается на левой границе интервала.

```python
n = np.arange(1,n+1)
M_1 = abs(np.sum(a_coef[1:]*n*a_s[:-1]))
dI = (b-a)*h/2*M_1

dI
```

Получили огроменную оценку погрешности, сравнимую по порядку с самим интегралом (напомним, он получился $\approx 37$). 

Такая погрешность связана с видом интегрируемой функции: по графику видно, что $M_1$ получается очень большим, а кроме того по условию шаг интегрирования не самый маленький ($h=0.1$)

В то же время посчитанное значение интеграла близко к аналитическому опять же из-за характера поведения функции: сначала она убывает, а потом возрастает, и таким образом части прямоугольников "торчащие" над функцией и пустоты "не достающие" до функции компенсируют друг друга, приводя к хорошему результату.

# Задание 4

```python
h = 0.1*2 # удваиваем шаг интегрирования для оценки посчитанного с шагом h=0.1 интеграла

xs = np.linspace(a,b, int((b-a)/h)+1)

ys = np.vander(xs, N=10, increasing=True) @ a_coef

I_2h = np.sum(ys[1:])*h

p = 1 # Порядок сходимости метода прямоугольников

dI_runge = abs((I_2h-I_chisl)/(2**p-1))

dI_runge
```

Видим, что оценка погрешности по методу Рунге дала гораздо меньшее число, чем аналитическая формула.

Проверим условия применения метода Рунге:

```python
def integr(a,b,h):
    xs = np.linspace(a,b, int((b-a)/h)+1)
    ys = np.vander(xs, N=10, increasing=True) @ a_coef
    return np.sum(ys[1:])*h 


def runge_checks(a, b, h0, eps=1e-12):
    h=h0
    Ih = integr(a,b,h0)
    Dh = None
    while True:
        I2h = integr(a,b,h*2)
        h = h/2
        Ih_2 = integr(a,b,h)
        Dh_2 = abs((Ih - Ih_2) / (2**1 - 1)) # p = 1 для метода прямоугольников
        Ch_2 = Dh_2 / h**1; ps = np.log2(Dh / Dh_2) if Dh != None else np.nan
        emp = abs(2**1*(Ih_2-Ih)/(Ih-I2h)-1)
        print('I(h) = %.16f, err(h) = %.6e, p* = %4.2f, C = %.6e, emp = %4.2f' % \
              (Ih_2, Dh_2, ps, Ch_2, emp))
        if abs(Dh_2) < eps:
            break
        I2h=Ih; Ih = Ih_2; Dh = Dh_2
        if h <0.000001: print('Too large n'); break
    return 
```

```python
runge_checks(a, b, 0.5)
```

Видим, что, строго говоря, критерии не выполнены, однако в то же время заметим, что 

$C_h$ неизменна **по порядку**, 

$p*$ изменяется в некоторой окрестности единицы (не принимая другие целые значения), 

а эмпирическое правило выполняется с некоторой периодичностью, однако значение **по порядку** всего близко к искомым $0.1$.

```python
Ih = integr(a,b,0.1) # Первоначально найденный интеграл, его и будем экстраполировать
Ih2 = integr(a,b,0.2) # В нашем случае в формуле экстраполяции h/2=0.1 - меньший шаг, h=0.2 - больший шаг

I_new = Ih+(Ih-Ih2)/(2**p-1)

print(f'Точное значение: I = {I_real},\nчисленно посчитанное: Ih = {I_chisl}, \nс помощью экстраполяции Ричардсона I_new = {I_new}')
```

# Задание 3

Интерполяционный многочлен Лагранжа выглядит следующим образом: $$L_n(x) = \sum_{i=1}^ny_il_i(x),$$ где $y_i$ - значения функции в точках $x_i$, а множители $l_i(x)$ задаются следующим образом: $$l_i(x) = \prod_{j=0, j\neq i}^n \dfrac{x-x_j}{x_i-x_j}$$

Видно, что по построению он будет проходить через все заданные точки таблицы, чего нам и нужно.

```python
# Часть таблицы без пропусков - по ней будем строить многочлен
table_wo_empty = np.array([[-0.8,-0.2,0,0.2,0.4,0.8,0.9,1],[-3.027,-0.206,1,1.406,1.504,1.904,2.392,3.2]])
```

Коэффициенты многочлена будем считать аналогично Заданию 1 - методом обратной матрицы:

```python
X = np.vander(table_wo_empty[0], increasing=True)
X_rev = np.linalg.inv(X)

# Находим столбец коэффициентов
a_coef = X_rev @ table_wo_empty[1]

np.around(a_coef,3)
```

Тогда сам многочлен будет выглядить следующим образом: $$L_8(x) = 1 + 3.498x-9.812x^2+13.946x^3-5.787x^4-15.913x^5+27.04x^6-10.773x^7$$

```python
xs = np.linspace(-0.8,1, 1000)
X_plt = np.vander(xs, 8, increasing=True)

ys = X_plt @ a_coef

plt.plot(xs, ys, color='navy', label="interpolation")
plt.scatter(table_wo_empty[0], table_wo_empty[1], color='crimson', label='data')
plt.legend()
plt.show()
```

Теперь заполним пропуски

```python
x_miss = -0.5

y_miss = np.vander([x_miss], 8, increasing=True) @ a_coef

plt.plot(xs, ys, color='navy', label="interpolation")
plt.scatter(table_wo_empty[0], table_wo_empty[1], color='crimson', label='data')
plt.scatter(x_miss, y_miss, color="green", label="missing y value")
plt.legend()
plt.show()
```

Теперь найдём пропущенный x:

```python
from scipy.optimize import fsolve # Находит корни функции

def f(x, a):
    xs = np.array([x**i for i in range(8)]).reshape(-1)
    return np.sum(np.dot(xs, a_coef))-a

y_for_x_missing = 1.644
x_missing = fsolve(f, 0.6, 1.644)

plt.plot(xs, ys, color='navy', label="interpolation")
plt.scatter(table_wo_empty[0], table_wo_empty[1], color='crimson', label='data')
plt.scatter(x_miss, y_miss, color="green", label="missing y value")
plt.scatter(x_missing, y_for_x_missing, color="cyan", label="missing x value")
plt.legend()
plt.show()
```
